# 3. Descriptive Analysis
Calculate and analyze basic network metrics such as node and edge counts, average degree, density, and more.

## 3.1 Calculate Various Network Metrics

```python
num_nodes = G.number_of_nodes()   # Node count

num_edges = G.number_of_edges()   # Edge count

# Average degree
average_degree = sum(dict(G.degree()).values()) / num_nodes

# Density
density = nx.density(G)

# Diameter (only if the graph is connected)
if nx.is_connected(G):
    diameter = nx.diameter(G)
else:
    diameter = None

# Clustering coefficient
clustering_coefficient = nx.average_clustering(G)
```

## 3.2 Display Metrics

```python
print(f"Number of nodes: {num_nodes}")
print(f"Number of edges: {num_edges}")
print(f"Average degree: {average_degree}")
print(f"Density: {density}")
print(f"Diameter: {diameter}")
print(f"Clustering coefficient: {clustering_coefficient}")
```

# 4. Centrality Measures + Visualization
Compute and interpret centrality measures like degree centrality, betweenness centrality, and eigenvector centrality to identify important nodes within the network.

## 4.1 Degree Centrality
Degree centrality measures how well-connected a node is by counting the number of its neighbors (edges connected to it).

```python
degree_centrality = nx.degree_centrality(G)

influential_nodes_degree = sorted(degree_centrality, key=degree_centrality.get, reverse=True)[:5]

# Printing the top 5 influential nodes and their degree centrality values
for node in influential_nodes_degree:
    centrality = degree_centrality[node]
    print(f"Node {node}: Degree Centrality = {centrality:.3f}")

# Printing all
# for node, centrality in degree_centrality.items():
#     print(f"Node {node}: Degree Centrality = {centrality:.3f}")
```

## 4.2 Betweenness Centrality
Betweenness centrality identifies nodes that act as bridges between different parts of the network. It measures the fraction of shortest paths that pass through a node.

```python
betweenness_centrality = nx.betweenness_centrality(G)

influential_nodes_betweenness = sorted(betweenness_centrality, key=betweenness_centrality.get, reverse=True)[:5]

# Printing the top 5 influential nodes and their betweenness centrality values
for node in influential_nodes_betweenness:
    centrality = betweenness_centrality[node]
    print(f"Node {node}: Betweenness Centrality = {centrality:.3f}")

# Printing all
# for node, centrality in betweenness_centrality.items():
#     print(f"Node {node}: Betweenness Centrality = {centrality:.3f}")
```

## 4.3 Eigenvector Centrality
Eigenvector centrality measures the influence of a node in the network, taking into account the centrality of its neighbors. It assumes that nodes connected to other central nodes are themselves central.

```python
eigenvector_centrality = nx.eigenvector_centrality(G)

influential_nodes_eigenvector = sorted(eigenvector_centrality, key=eigenvector_centrality.get, reverse=True)[:5]

# Printing the top 5 influential nodes and their eigenvector centrality values
for node in influential_nodes_eigenvector:
    centrality = eigenvector_centrality[node]
    print(f"Node {node}: Eigenvector Centrality = {centrality:.3f}")

# Printing all
# for node, centrality in eigenvector_centrality.items():
#     print(f"Node {node}: Eigenvector Centrality = {centrality:.3f}")
```

## 4.4 Closeness Centrality
Closeness centrality measures how close a node is to all other nodes in the network. It's calculated as the reciprocal of the sum of the shortest path distances from a node to all other nodes. Nodes with high closeness centrality can quickly reach other nodes in the network.

```python
closeness_centrality = nx.closeness_centrality(G)

influential_nodes_closeness = sorted(closeness_centrality, key=closeness_centrality.get, reverse=True)[:5]

# Printing the top 5 influential nodes and their closeness centrality values
for node in influential_nodes_closeness:
    centrality = closeness_centrality[node]
    print(f"Node {node}: Closeness Centrality = {centrality:.3f}")

# Printing all
# for node, centrality in closeness_centrality.items():
#     print(f"Node {node}: Closeness Centrality = {centrality:.3f}")
```

## 4.5 Load Centrality
Load centrality measures the importance of a node based on the concept that nodes carrying more traffic or information are more central. It's commonly used in transportation and communication networks.

```python
load_centrality = nx.load_centrality(G)

influential_nodes_load = sorted(load_centrality, key=load_centrality.get, reverse=True)[:5]

# Printing the top 5 influential nodes and their load centrality values
for node in influential_nodes_load:
    centrality = load_centrality[node]
    print(f"Node {node}: Load Centrality = {centrality:.3f}")

# Printing all
# for node, centrality in load_centrality.items():
#     print(f"Node {node}: Load Centrality = {centrality:.3f}")
```

# 5. Connectivity Analysis
Explore network connectivity by identifying connected components, isolates, and articulation points.

## 5.1 Connected Components
Connected components are groups of nodes in a network that are connected to each other but not connected to nodes outside the group. You can find connected components using the 'nx.connected_components' function.

```python
# Detect connected components
connected_components = list(nx.connected_components(G))

# Print the connected components
count = 0
for i, component in enumerate(connected_components):
    count+=1
    print(f"Connected Component {i + 1}: {component}")
print('No. of connected components:',count)
```

## 5.2 Isolates
Isolates are nodes that are not connected to any other nodes in the network. You can find isolates using a list comprehension to iterate through the nodes and check their degree. Nodes with a degree of 0 are isolates.

```python
isolates = [node for node in G.nodes if G.degree(node) == 0]

# Print isolates
print("Isolates:", isolates)
```

## 5.3 Articulation Points (Cut Vertices)
Articulation points are nodes whose removal would increase the number of connected components in the network. You can find articulation points using the 'nx.articulation_points' function.

```python
articulation_points = list(nx.articulation_points(G))

# Print articulation points
print("Articulation Points:", articulation_points)
```

## 5.4 Removing Articulation Points
Removing articulation points of the network to observe if the no. of connected components increases or not.

```python
# Creating a copy of graph to not change the original graph
H = G

# Removing the articulation points
H.remove_nodes_from(articulation_points)
```

### 5.4.1 Recalculating the connected components

```python
connected_components = list(nx.connected_components(H))
count = 0
for i, component in enumerate(connected_components):
    count+=1
    print(f"Connected Component {i + 1}: {component}")
print('No. of connected components:',count)
```

### 5.4.2 Recalculating isolates

```python
isolates = [node for node in G.nodes if G.degree(node) == 0]
print("Isolates:", isolates)
```

### 5.4.3 Visualizing the new graph

```python
nx.draw(H,pos,with_labels=True,node_color='pink',node_size=600,font_size=7,edge_color='red')
plt.title('New Dolphins Social Network Graph')
plt.show()
```

# 6. Community Detection + Visualization
Apply community detection algorithms (e.g., Louvain, Girvan-Newman) to uncover clusters or communities within the network. Different communities detected are visualized using different colors.

## 6.1 Louvain Community Detection
Louvain is a popular algorithm for detecting communities in networks. To use it with NetworkX, you'll need to install the python-louvain package, which provides an implementation of the Louvain algorithm.

### 6.1.1 Applying the Louvain algorithm for community detection

```python
import community.community_louvain as community_louvain  # Import the community module from python-louvain

# Generate a dendrogram using the Louvain algorithm
dendrogram = community_louvain.generate_dendrogram(G)

# Get the final partition from the dendrogram
partition = community_louvain.partition_at_level(dendrogram, len(dendrogram) - 1)

# Print the detected communities
communities = {}
for node, community_id in partition.items():
    if community_id not in communities:
        communities[community_id] = []
    communities[community_id].append(node)

for community_id, nodes in communities.items():
    print(f"Community {community_id}: {nodes}")
```

### 6.1.2 Visualizing the communties detected using the Louvain algorithm

```python
import numpy as np

# Assign community membership as node attributes
membership = {}
for i, comm in enumerate(partition.items()):
    for node in comm:
        membership[node] = i

nx.set_node_attributes(G, membership, "community")

# Position the nodes using a spring layout (replace with your preferred layout)
pos = nx.kamada_kawai_layout(G)

# Create a list of node colors based on community membership
node_colors = [membership[node] for node in G.nodes()]

# Create the plot
plt.figure(figsize=(9,9))
nx.draw_networkx(G, pos, node_color=node_colors, cmap="tab10", with_labels=True, node_size=900, font_size=8,alpha=0.8)
# nx.draw_networkx(G, pos, node_color=node_colors, cmap="rainbow", with_labels=True, node_size=900, font_size=8,alpha=0.8)
# nx.draw_networkx(G, pos, node_color=node_colors, cmap="cool", with_labels=True, node_size=900, font_size=8,alpha=0.8)
nx.draw_networkx_labels(G, pos, font_size=8)
nx.draw_networkx_edges(G, pos)
plt.title("Community Detection with Louvain Algorithm")
plt.show()
```

## 6.2 Girvan-Newman Community Detection
The Girvan-Newman algorithm is used to detect communities by iteratively removing edges from the network. NetworkX provides functions to implement this algorithm.

### 6.2.1 Applying the Girvan-Newman algorithm for community detection

```python
# Apply Girvan-Newman algorithm to detect communities
communities_generator = nx.community.girvan_newman(G)
top_level_communities = next(communities_generator)

# Print the detected communities
for i, community in enumerate(top_level_communities):
    print(f"Community {i + 1}: {community}")
```

### 6.2.2 Visualizing the communties detected using the Girvan-Newman algorithm

```python
import numpy as np

# Assign community membership as node attributes
membership = {}
for i, comm in enumerate(top_level_communities):
    for node in comm:
        membership[node] = i

nx.set_node_attributes(G, membership, "community")

# Position the nodes using a spring layout (replace with your preferred layout)
pos = nx.kamada_kawai_layout(G)

# Create a list of node colors based on community membership
node_colors = [membership[node] for node in G.nodes()]

# Create the plot
plt.figure(figsize=(9,9))
# nx.draw_networkx(G, pos, node_color=node_colors, cmap="tab10", with_labels=True, node_size=900, font_size=8,alpha=0.8)
nx.draw_networkx(G, pos, node_color=node_colors, cmap="rainbow", with_labels=True, node_size=900, font_size=8,alpha=0.8)
# nx.draw_networkx(G, pos, node_color=node_colors, cmap="cool", with_labels=True, node_size=900, font_size=8,alpha=0.8)
nx.draw_networkx_labels(G, pos, font_size=8)
nx.draw_networkx_edges(G, pos)
plt.title("Community Detection with Girvan-Newman Algorithm")
plt.show()
```
